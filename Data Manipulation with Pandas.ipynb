```python
import pandas as pd
sales = pd.read_csv('Downloads/sales_subset.csv')
sales.head()
```

## Aggregating Data

```python
sales.info()
```

```python
# Print the mean of weekly sales
sales_mean = sales['weekly_sales'].mean()

sales_median = sales['weekly_sales'].median()

sales_median
```

```python
sales_mean
```

```python
sales['date'].max()
```

```python
## Efficient Summaries
```

```python
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)
sales['temperature_c'].agg(iqr)

sales[['temperature_c','fuel_price_usd_per_l', 'unemployment']].agg(iqr)
```

```python
import numpy as np
sales[['temperature_c','fuel_price_usd_per_l', 'unemployment']].agg([iqr, np.median])
```

```python
sales1_1 = sales.sort_values('date', ascending = True)
sales1_1
```

```python
sales1_1['cum_weekly_sales'] = sales['weekly_sales'].cumsum()
sales1_1.head()
```

```python
sales1_1['cum_max_sales'] = sales['weekly_sales'].cummax()
sales1_1.head()
```

```python
sales1_1[['date', 'cum_weekly_sales', 'cum_max_sales']]
```

```python
##Dropping Duplicates
```

```python
store_types = sales.drop_duplicates(subset=['store', 'type'])
store_types
```

```python
store_depts = sales.drop_duplicates(subset=['store', 'department'])
store_depts.head()
```

```python
## subset the holiday days and drop duplicate dates
```

```python
holiday_dates = sales[sales['is_holiday']==True].drop_duplicates('date')
holiday_dates['date']
```

```python
## Counting Categorical variables
store_counts = store_types['type'].value_counts()
store_counts
```

```python
## Get the proportion of each store type
store_p = store_types['type'].value_counts(normalize = True)
store_p
```

```python
dept_count_sort = store_depts['department'].value_counts(sort=True)
dept_count_sort
```

```python
dept_count_sort_prop = store_depts['department'].value_counts(sort = True, normalize = True)
dept_count_sort_prop
```

```python
sales_all = sales['weekly_sales'].sum()
sales_all
```

```python
sales_A = sales[sales['type'] =='A']['weekly_sales'].sum()
sales_A
sales_B = sales[sales['type'] =='B']['weekly_sales'].sum()
sales_B
sales_C = sales[sales['type'] == 'C']['weekly_sales'].sum()
sales_C
```

```python
sales_propn_type = [sales_A, sales_B, sales_C] / (sales_A + sales_B + sales_C)
sales_propn_type
```

```python
## Using Groupby
```

```python
sales_type = sales.groupby('type')['weekly_sales'].sum()
sales_type
```

```python
sales_type_propn = sales_type / sum(sales.weekly_sales)
sales_type_propn
```

```python
sales_by_type_isholiday = sales.groupby(['type','is_holiday'])['weekly_sales'].sum()
sales_by_type_isholiday
```

```python
## Multiple grouped Summaries
```

```python
import numpy as np

## for each store type, aggregate weekly sales: mean, median, min, max
sales_type_agg = sales.groupby('type')['weekly_sales'].agg([np.mean, np.max, np.min, np.median])
sales_type_agg
```

```python
## for each store type, aggregate unemployment, fuel_price_usd_l: mean, median, min, max
sales_type_agg_un_fu = sales.groupby('type')[['unemployment','fuel_price_usd_per_l']].agg([np.mean, np.median, np.min, np.max])
sales_type_agg_un_fu
```

```python
## Pivot table
```

```python
## Pivot for weekly sales by store type
sales_pi_type = sales.pivot_table(values = 'weekly_sales', index = 'type')
sales_pi_type
```

```python
## Pivot for mean median of weekly sales by store type
mean_med_sales_type = sales.pivot_table(values='weekly_sales', index = 'type', aggfunc=[np.mean, np.median])
mean_med_sales_type
```

```python
## Pivot for mean weekly sales by store type and is_holiday
mean_sales_is_h = sales.pivot_table(values = 'weekly_sales', index ='type', columns='is_holiday')
mean_sales_is_h
```

```python
## Fill in missing values and sum values of pivot table
```

```python
## mean weekly sales by department and type: fill missing values with 0
sales_mean = sales.pivot_table(values = 'weekly_sales', index = 'department', columns = 'type', fill_value = 0)
sales_mean
```

```python
## mean weekly sales by department and type: fill missing values with 0: sum all rows and col
sales_mean_mar = sales.pivot_table(values = 'weekly_sales', index='department', columns = 'type', fill_value = 0, margins = True)
sales_mean_mar
```
