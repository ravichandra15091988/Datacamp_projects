```python
import pandas as pd

countries = pd.read_csv('Downloads/countries-of-the-world.csv')
countries.head()

import seaborn as sns
import matplotlib.pyplot as plt

gdp = countries['GDP ($ per capita)']
phones = countries['Phones (per 1000)']
phones
percent_literate = countries['Literacy (%)']

sns.scatterplot(x = gdp, y = phones)

sns.scatterplot(x = gdp, y = percent_literate)
```

```python
region = countries['Region']

sns.countplot(y = region)

plt.show()
```

# Visualizing two Quantitative variables

```python
## Creating subplots with col and row

sns.relplot(x='absences', y ='G3', data= student_data, kind='scatter', col ='study_time')

sns.relplot(x='absences', y ='G3', data = student_data, kind= 'scatter', row = 'study_time')
```

```python
sns.relplot(x='G1', y ='G3', data = student_data, kind ='scatter', col = 'schoolsup', col_order = ['yes','no'])

sns.relplot(x='G1', y ='G3', data =student_data, kind ='scatter', col = 'schoolsup', col_order =['yes','no'], row = 'famsup', 
            row_order = ['yes','no'])
```

```python
## Changing the size of scatter plot points

mpg = pd.read_csv('Downloads/mpg.csv')

mpg
```

```python
sns.relplot(x = 'horsepower', y = 'mpg', data =mpg, kind = 'scatter', size = 'cylinders', hue ='cylinders')
```

```python
## Changing the style of scatterplot points

sns.relplot(x ='acceleration', y ='mpg', data = mpg, kind = 'scatter', size= 'origin', hue = 
        'origin')
```

```python
## Interpreting line plots

sns.relplot(x = 'model_year', y = 'mpg', data = mpg, kind ='line', errorbar='sd')
```

```python
## Plotting subgroups in lineplots

sns.relplot(x = 'model_year', y ='horsepower', data = mpg, kind ='line', errorbar = None, style ='origin', hue='origin', 
            markers = True, dashes = False)
```

## Visualizing categorical and quantitative variable



```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
survey_data = pd.read_csv('Downloads/young-people-survey-responses.csv', index_col =0)


survey_data['Age_category'] = np.where(survey_data['Age'] < 21, 'less than 21', '21+')

survey_data.head()
```

```python
sns.catplot(y = 'Internet usage', data = survey_data, kind = 'count')
```

```python
sns.catplot(y ='Internet usage', data = survey_data, kind = 'count', col ='Age_category')
```

```python
survey = pd.read_csv('Downloads/survey_data.csv')
survey.head()
```

```python
sns.catplot(x = 'Gender', y = 'Interested in Math', data = survey, kind ='bar')
```

## Customizing Bar plots

```python
sns.catplot(x = 'study_time', y = 'G3', data = student_data, kind='bar')

category_order = ['<2 hours', '2 to 5 hours', ' 5 to 10 hours', '>10 hours']

sns.catplot(x = 'study_time', y ='G3', data = student_data, kind = 'bar', order = category_order, errorbar = None)
```

## Box plots

```python
## Create box plots

study_time_order = ['<2 hours', '2 to 5 hours', ' 5 to 10 hours', '>10 hours']

sns.catplot(x= 'study_time', y= 'G3', data = student_data, kind = 'box', order = study_time_order)
```

## Customizing Seaborn plots

## Adding Titles and labels

```python
g = sns.relplot(x = 'weight', y = 'horsepower', data = mpg, kind ='scatter')

type_of_g = type(g)
type_of_g

g.fig.suptitle("Car Weight vs Horsepower")
```

## Adding a title and axis label

```python
mpg_mean = mpg.groupby(['model_year', 'origin']).agg(mpg_mean=('mpg','mean')).reset_index()

g = sns.lineplot(x = 'model_year', y = 'mpg_mean', hue='origin', data = mpg_mean)

g.set_title('Avg Mean by model year')

g.set(xlabel = 'Car Model Year',
            ylabel= 'Average MPG')
```
