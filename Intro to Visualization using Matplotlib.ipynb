## Plotting Timeseries data

```python
import pandas as pd

climate_change = pd.read_csv('Downloads/climate_change.csv', parse_dates = ['date'], index_col = 'date')
climate_change
```

```python
import matplotlib.pyplot as plt
fig, ax= plt.subplots()

ax.plot(climate_change.index, climate_change['relative_temp'])

ax.set_xlabel('Time')
ax.set_ylabel('Relative temperature (Celsius)')

plt.show()
```

```python
fig, ax = plt.subplots()

seventies = climate_change["1970-01-01":"1979-12-31"]
print(seventies)

ax.plot(seventies.index, seventies['co2'])

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(climate_change.index, climate_change['co2'], color = 'blue')
ax2 = ax.twinx()

ax2.plot(climate_change.index, climate_change['relative_temp'], color = 'red')

plt.show()
```

```python
## Defing a function that plots the timeseries data

def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color = color)
    axes.tick_params('y', colors=color)
```

```python
## Using timeseries function

fig, ax = plt.subplots()

plot_timeseries(ax, climate_change.index, climate_change['co2'], 'blue', 'Time (Years)', 'Co2 levels')

ax2 = ax.twinx()

plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', 'Time (Years)', "Relative temperature (Celsius)")
plt.show()
```

```python
## Annotatting a timeseries plot
fig, ax = plt.subplots()

ax.plot(climate_change.index, climate_change['relative_temp'])

ax.annotate('> 1 degree celsius', xy = (pd.Timestamp('2015-10-06'),1))

plt.show()
```

```python
## Putting it altogether

fig, ax = plt.subplots()

plot_timeseries(ax, climate_change.index, climate_change['co2'], 'blue', 'Time (Years)', 'Co2 levels')

ax2 = ax.twinx()

plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', 'Time (years)', 'relative temperature (degree)')

ax2.annotate('>1 degree', xy = (pd.Timestamp('2015-10-06'), 1), xytext = (pd.Timestamp('2008-10-06'), -0.2,), 
             arrowprops = {'arrowstyle':'->', 'color':'green'})
plt.show()
```

# Quantitative Comparisons: Bar charts

```python
medals = pd.read_csv('Downloads/medals_by_country_2016.csv', index_col = 0)

fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'])

ax.set_xticklabels(medals.index, rotation = 90)

ax.set_ylabel('Number of Medals')

plt.show()
```

## Stacked Bar chart

```python
fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'], label = 'Gold')

ax.bar(medals.index, medals['Silver'], bottom=medals['Gold'], label = 'Silver')

ax.bar(medals.index, medals['Bronze'], bottom=medals['Gold'] + medals['Silver'], label = 'Bronze')

ax.set_xticklabels(medals.index, rotation = 90)

ax.legend()
plt.show()
```

## Histogram

```python
summer_2016 = pd.read_csv('Downloads/summer2016.csv', index_col = 0)

mens_rowing = summer_2016[(summer_2016['Sport'] == 'Rowing') & (summer_2016['Sex'] == 'M')]
mens_rowing

mens_gymnastics = summer_2016[(summer_2016['Sport'] == 'Gymnastics') & (summer_2016['Sex']=='M')]
mens_gymnastics

fig, ax = plt.subplots()

ax.hist(mens_rowing['Weight'])

ax.hist(mens_gymnastics['Weight'])

ax.set_xlabel('Weight (Kg)')
ax.set_ylabel('# of Observations')

plt.show()
```

```python
## Step Histogram

fig, ax = plt.subplots()

ax.hist(mens_rowing['Weight'], histtype = 'step', label ='Rowing', bins = 5)

ax.hist(mens_gymnastics['Weight'], histtype = 'step', label = 'Gymnastics', bins = 5)

ax.set_xlabel('Weight (kg)')
ax.set_ylabel('# of Obesrvations')

ax.legend()
plt.show()
```

```python
## Adding error bars to a bar chart

fig, ax = plt.subplots()

ax.bar('Rowing', mens_rowing['Height'].mean(), yerr = mens_rowing['Height'].std())

ax.bar('Gymnastics', mens_gymnastics['Height'].mean(), yerr = mens_gymnastics['Height'].std())

ax.set_ylabel('Height (cm)')

plt.show()
```

```python
## Adding error bars to a plot

fig, ax = plt.subplots()

ax.errorbar(seatle_weather['MONTH'], seatle_weather['MLY-TAVG-NORMAL'], yerr = seatle_weather['MLY-TAVG-STDDEV'])

ax.errorbar(austin_weather['MONTH'], austin_weather['MLY-TAVG-NORMAL'], yerr= austin_weather['MLY-TAVG-STDDEV'])

ax.set_ylabel('Temperature (Fahrenheit)')

plt.show()
```

## Creating Boxplots

```python
fig, ax = plt.subplots()

ax.boxplot([mens_rowing['Height'], mens_gymnastics['Height']])

ax.set_xticklabels(['Rowing', 'Gymnastics'])

ax.set_ylabel('Height (cm)')

plt.show()
```
